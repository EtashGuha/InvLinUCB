```python
def LinUCB(V,arm,reward,A_t,y,beta):
    
    # estimate theta 
    V = V + arm[A_t].T * arm[A_t]
    breakpoint()
    y = y + arm[A_t].T * reward
    inv_V = np.linalg.inv(V)
    print(inv_V.shape)
    breakpoint()
    hat_theta = inv_V * y
    
    # compute ucb
    ucb = arm * hat_theta + beta * np.matrix(np.diag(np.sqrt( arm * inv_V * arm.T ))).T

    print(ucb.shape)
    return np.argmax(ucb),V,y


def UCB(hat_mu,reward,A_t,N,delta,k,i):
    
    if k < i-1:
        hat_mu[A_t] = hat_mu[A_t] + reward
        A_next = k + 1
    else:
        hat_mu[A_t] = hat_mu[A_t] + reward
        ucb = hat_mu/N + np.sqrt((2*np.log(1/delta))/N) 
        A_next = np.argmax(ucb)

    return A_next,hat_mu


#parameters of problem 
d = 2
T = 1000
rep = 200
number_of_arms = 2
number_of_Deltas = 20



#LinUCB 
beta = 1 + np.sqrt(2*np.log(T)+d*np.log((d+T)/d))
lamda = 1
arm = np.matrix([[1,0],[0,1]]) # set of arms
breakpoint()
reg = np.zeros([number_of_Deltas,rep]) # regret under each delta and repeats




count = 0
for i in np.arange(0,1,1/number_of_Deltas)+ 1/number_of_Deltas:
    print(i)
    mu = np.matrix([[0,-i]]).T # mean of arms
    
    for j in range(0,rep,1):
        
        #initialization
        V = lamda * np.identity(d)
        hat_theta = np.zeros([d,1])
        A_t = np.random.randint(number_of_arms)
        y = np.zeros([d,1])
        N = np.zeros([number_of_arms,1])
        
        for k in range(0,T,1):
            N[A_t] = N[A_t] + 1 # pull an arm
            reward = mu.T * arm[A_t].T + np.random.normal(0,1) # observe reward
            A_t,V,y = LinUCB(V,arm,reward,A_t,y,beta) # update decision
            break
        break
        reg[count,j] = N[1] * i # compute regret
    count = count + 1

        
fin_LinUCB = (np.sum(reg,axis=1)) / rep


#UCB 
delta = 1/T**2
arm = np.matrix([[0,1]])
reg = np.zeros([number_of_Deltas,rep])

count = 0
for i in np.arange(0,1,1/number_of_Deltas)+ 1/number_of_Deltas:
    
    mu = np.matrix([[0,-i]]).T
    
    for j in range(0,rep,1):
        
        #initialization
        A_t = 0
        hat_mu = np.zeros([number_of_arms,1])
        N = np.zeros([number_of_arms,1])
        
        for k in range(0,T,1):
            N[A_t] = N[A_t] + 1
            reward = mu[A_t] + np.random.normal(0,1)
            A_t,hat_mu = UCB(hat_mu,reward,A_t,N,delta,k,2)
            
        break
        
        reg[count,j] = N[1] * i
    break
    count = count + 1

    
fin_UCB = (np.sum(reg,axis=1)) / rep


#plot
Delta = np.arange(0,1,1/number_of_Deltas) + 1/number_of_Deltas
plt.plot(Delta, fin_LinUCB,label="LinUCB")
plt.plot(Delta, fin_UCB,label="UCB")
plt.xlabel("Delta")
plt.ylabel("Regret")
plt.legend()
plt.grid()
plt.show()
```

```python
def find_gap(arm,theta):   
    return np.max(arm*theta)-arm*theta

def sample_spherical(npoints, ndim=5):
    vec = np.random.randn(ndim, npoints)
    vec /= np.linalg.norm(vec, axis=0)
    return vec

def LinUCB(V,arm,reward,A_t,y,beta):
    
    V = V + arm[A_t].T * arm[A_t]
    y = y + arm[A_t].T * reward
    inv_V = np.linalg.inv(V)
    hat_theta = inv_V * y

    ucb = arm * hat_theta + beta * np.matrix(np.sqrt(np.diag( arm * inv_V * arm.T ))).T


    return np.argmax(ucb),V,y


def UCB(hat_mu,reward,A_t,N,delta,k,i):
    
    if k < i-1:
        hat_mu[A_t] = hat_mu[A_t] + reward
        A_next = k + 1
    else:
        hat_mu[A_t] = hat_mu[A_t] + reward
        ucb = hat_mu/N + np.sqrt((2*np.log(1/delta))/N) 
        A_next = np.argmax(ucb)

    return A_next,hat_mu




#parameters of problem 
d = 5
theta = np.sqrt(1/30) * np.matrix([[4,3,2,1,0]]).T

arms_set = np.matrix(sample_spherical(1000)).T

T = 5000
rep = 100
number_of_arms = 1000

sample = 100
sample_K = np.size(np.arange(2,1000,sample))


#LinUCB 
beta = 1 + np.sqrt(2*np.log(T)+d*np.log((d+T)/d))
lamda = 1
reg = np.zeros([sample_K,rep])


count = 0
for i in np.arange(2,number_of_arms,sample):
    arm = arms_set[0:i,:]
    gap = find_gap(arm,theta)
    

    for j in range(0,rep,1):
        
        #initialization
        V = lamda * np.identity(d)
        hat_theta = np.zeros([d,1])
        A_t = np.random.randint(i)
        y = np.zeros([d,1])
        N = np.zeros([i,1])
        
        for k in range(0,T,1):
            N[A_t] = N[A_t] + 1
            reward = theta.T * arm[A_t].T + np.random.normal(0,1)
            A_t,V,y = LinUCB(V,arm,reward,A_t,y,beta)
        reg[count,j] = N.T * gap
    count = count + 1

        
fin_LinUCB = (np.sum(reg,axis=1)) / rep



#UCB 
delta = 1/T**2

reg = np.zeros([sample_K,rep])

count = 0
for i in np.arange(2,1000,sample):
    arm = arms_set[0:i,:]
    gap = find_gap(arm,theta)
    
    for j in range(0,rep,1):
        
        #initialization
        A_t = 0
        hat_mu = np.zeros([i,1])
        N = np.zeros([i,1])
        
        for k in range(0,T,1):
            N[A_t] = N[A_t] + 1
            reward = theta.T * arm[A_t].T + np.random.normal(0,1)
            A_t,hat_mu = UCB(hat_mu,reward,A_t,N,delta,k,i)
        reg[count,j] = N.T * gap
    count = count + 1

fin_UCB = (np.sum(reg,axis=1)) / rep


K = np.arange(2,1000,sample)
plt.plot(K, fin_LinUCB,label="LinUCB")
plt.plot(K, fin_UCB,label="UCB")
plt.xlabel("K")
plt.ylabel("Regret")
plt.legend()
plt.grid()
plt.show()
```
